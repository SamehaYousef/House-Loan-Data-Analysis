```python
def fill_mid(col, df):
    for c in col:
        print("type1", new_df[c].dtype==object)


        if new_df[c].dtypes == object:
            freq = new_df[c].value_counts().index[0]
            num = new_df[c].value_counts().iloc[0]
            print("freq", freq)
            
            print("nas",new_df[c].isna().sum())
            print("num", num)
            df[[c]] = df[[c]].fillna(freq)
            
            num = new_df[c].value_counts()[0]

            print("num", num)




        
        else:    
            print(c)
            mid = new_df[c].median()
            print(mid)
            df[[c]] = df[[c]].fillna(mid)
    return df


new_df = fill_mid(adj.index, new_df)
```

```python
from sklearn.metrics import classification_report
```

```python
model = Sequential()

model.add(Dense(256, input_dim = 158, activation = 'relu'))
model.add(Dropout(0.2)),
model.add(Dense(256,                   activation = 'elu'))
model.add(Dropout(0.3)),
model.add(Dense(128,                   activation = 'relu'))
model.add(Dense(64,                   activation = 'elu'))
model.add(Dropout(0.2)),
model.add(Dense(12,                   activation = 'relu'))
model.add(Dropout(0.1)),
model.add(Dense(8,                   activation = 'elu'))
model.add(Dense(4,                   activation = 'relu'))
model.add(Dense(1,                 activation='sigmoid'))
```
